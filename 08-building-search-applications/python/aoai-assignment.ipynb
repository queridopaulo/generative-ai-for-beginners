```python
import os
import numpy as np
from ollama import Client
from dotenv import load_dotenv
load_dotenv()

client = Client(
    host = os.environ['OLLAMA_ENDPOINT']
    )
model = "mistral" # Change the model accordingly
```

```python
# Dependencies for embeddings_utils
%pip install numpy matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'

result = client.embeddings(model=model, prompt = text)

result['embedding']
```

```python
# compare several wordd
automobile_embedding    = client.embeddings(model=model, prompt = 'automobile')
vehicle_embedding       = client.embeddings(model=model, prompt = 'vehicle')
dinosaur_embedding      = client.embeddings(model=model, prompt = 'dinosaur')
stick_embedding         = client.embeddings(model=model, prompt = 'stick')

# comparing cosine similarity, automobiles vs automobiles should be 1.0, i.e exactly the same, while automobiles vs dinosaurs should be between 0 and 1, i.e. not the same
print(cosine_similarity(automobile_embedding['embedding'], automobile_embedding['embedding']))
print(cosine_similarity(automobile_embedding['embedding'], vehicle_embedding['embedding']))
print(cosine_similarity(automobile_embedding['embedding'], dinosaur_embedding['embedding']))
print(cosine_similarity(automobile_embedding['embedding'], stick_embedding['embedding']))
```
